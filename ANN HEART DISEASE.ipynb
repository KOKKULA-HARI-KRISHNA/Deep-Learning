```python
import pandas as pd
import numpy as np
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ANN/ann_HeartDisease.csv')
```

```python
data
```

```python
data = data.drop('Unnamed: 0', axis = 1)
```

```python
data.shape
```

```python
x = data.drop('target', axis = 1)
y = pd.DataFrame(data['target'], columns=['target'])
```

```python
from imblearn.over_sampling import RandomOverSampler
random = RandomOverSampler(random_state = 1)
x_sampled,y_sampled = random.fit_resample(x,y)
```

```python
x_sampled.shape ,y_sampled.shape
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_sampled,y_sampled, test_size = 0.3,random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train.shape
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,ReLU
from keras.layers import Dropout
```

```python
classifier = Sequential()
classifier.add(Dense(units= 32,activation='relu',kernel_initializer='he_uniform',input_dim = 22))
classifier.add(Dense(units= 16,activation='relu',kernel_initializer='he_uniform'))
classifier.add(Dense(units= 8,activation='relu',kernel_initializer='he_uniform'))
classifier.add(Dense(units= 1,kernel_initializer='glorot_uniform', activation='sigmoid'))
```

```python
classifier.compile(optimizer='Adam',loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
model_history=classifier.fit(x_train, y_train,validation_split=0.33, epochs=2000, batch_size = 50)
```

```python
y_predict_prob = classifier.predict(x_test)
```

```python
y_predict = []
for i in y_predict_prob:
  if i> 0.7669565217391304:
    y_predict.append(1)
  else:
    y_predict.append(0)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test,y_predict)
```
